### 15.Replace all the null values in ReadmissionFlag with 0.

```python
import pandas as pd
import numpy as np
```

```python
workbook_url = 'HospitalDatabase.xlsx'
ReAdmissionRegistry_df = pd.read_excel(workbook_url,sheet_name='ReAdmissionRegistry')
ReAdmissionRegistry_df
```

```python
ReAdmissionRegistry_df['ReadmissionFlag'].isna().sum()
```

```python
ReAdmissionRegistry_df['ReadmissionFlag'] = ReAdmissionRegistry_df['ReadmissionFlag'].fillna(0).inplace=True
```
