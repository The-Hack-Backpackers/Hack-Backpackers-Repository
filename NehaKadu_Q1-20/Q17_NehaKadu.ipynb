### 17.Create a heat map to show the correlation between the column values of Edvisits.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
workbook_url = 'HospitalDatabase.xlsx'
EDVisits_df = pd.read_excel(workbook_url,sheet_name='EDVisits')
EDVisits_df
```

```python
plt.figure(figsize=(16, 6))
correlation = sns.heatmap(EDVisits_df.corr(), annot=True)
```
