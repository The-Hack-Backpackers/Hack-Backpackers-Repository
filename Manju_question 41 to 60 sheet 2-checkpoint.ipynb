# 45 Calculate mortality by diagnosis.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('HospitalDatabase-Discharges.csv')
pd.pivot_table(df,index=['PrimaryDiagnosis','ExpectedMortality'])
```
